__Load data and Primary analysis__

```python
#Loading the data and making a copy of each of them
print('Test data')
test_dataset= pd.read_csv('.\data\Test.csv')
test_df=test_dataset.copy()
display(test_df.head(5))

print('Train data')
train_dataset= pd.read_csv('.\data\Train.csv')
train_df= train_dataset.copy()
display(train_df.head(5))


print('Riders details')
rider_dataset= pd.read_csv('.\data\Riders.csv')
rider_df=rider_dataset.copy()
print(rider_df.head(5))
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
#checking the (target columns), columns that have been dropped in test_df
for col in train_df.columns:
    if col not in test_df.columns:
        print(col)
```

```python
#checking for missing values.
#Temparature and Precipitation in millimeters columns have missing values.

train_null=train_df.isnull().sum()
print(train_null[train_null>0])

test_null=test_df.isnull().sum()
print(test_null[test_null>0])


```

```python
#check the percentage of missing values as a %of the total.
null_train=train_df.isnull().sum()[train_df.isnull().sum()>0]
null_train_df=pd.DataFrame({'NaN_count':null_train,'NaN_%':(null_train/len(train_df))})
display(null_train_df)

null_test=test_df.isnull().sum()[train_df.isnull().sum()>0]
null_test_df=pd.DataFrame({'NaN_count':null_test,'NaN_%':(null_test/len(test_df))})
display(null_test_df)
```

__Notes__
- Around 20% of temparature values are missing.
- Around 97% of Precipitation in Millimeters values are missing.
