```python
#check the percentage of missing values as a %of the total.
def missing(dataset):
    nan_count= dataset.isnull().sum()[dataset.isnull().sum()>0]
    nan_df=pd.DataFrame({'NaN_count': nan_count,'Nan_%':(nan_count/len(dataset))})
    
    print(f'Missing Percentage in')
    display(nan_df)


missing(train_df)
missing(test_df)
```
